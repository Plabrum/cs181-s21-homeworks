```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook
```

```python
decay_score = np.load("decay_score.npy")
no_decay_score = np.load("no_decay_score.npy")
```

```python
plt.figure(figsize=(8,6))
plt.plot(decay_score)
plt.title("Score vs Epoch with Epsilon Decay")
plt.ylabel("Score")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(no_decay_score)
plt.title("Score vs Epoch with no Epsilon Decay")
plt.ylabel("Score")
plt.xlabel("Epoch")
plt.show()
```

```python
a_g_score = np.load("a_g_score.npy")
```

```python
a_g_score
```

```python
df = pd.DataFrame(a_g_score)
df.columns = ["Alpha", "Gamma", "Score"]
df
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
ax.set_xlabel('Alpha')
ax.set_ylabel('Gamma')
ax.set_zlabel('Score')
ax.axes.set_xticks([0,1,2,3])
ax.axes.set_yticks([0,1,2,3])
ax.axes.set_xticklabels([0.001, 0.01, 0.1, 0.2])
ax.axes.set_yticklabels([0.001, 0.01, 0.1, 0.2])
surf=ax.plot_trisurf(df['Alpha'], df['Gamma'], df['Score'], cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
